# Load Dataset, Eksplorasi Data Sederhana, Data Cleaning

Dataset dapat diunduh di https://www.kaggle.com/andrewmvd/heart-failure-clinical-data

Model machine learning jenis klasifikasi ini dibuat untuk memprediksi kemungkinan kematian akibat gagal jantung dari data kesehatan pasien.

Kolom:
- age -> usia pasien
- anaemia -> Apakah pasien menderita anaemia? (Ya:1 dan No:0)
- creatinine_phosphokinase -> Level of the CPK enzyme in the blood (mcg/L)
- diabetes -> ada diabetes? (Ya:1 dan No:0)
- ejection_fraction -> Percentage of blood leaving the heart at each contraction (percentage)
- high_blood_pressure -> Apakah pasien memiliki hipertensi? (Ya:1 dan No:0)
- plateles -> Platelets in the blood (kiloplatelets/mL)
- serum_creatinine -> Level of serum creatinine in the blood (mg/dL)
- serum_sodium -> Level of serum sodium in the blood (mEq/L)
- sex -> kelamin (Pria:1, Wanita:0)
- smoking -> Pasien perokok? (Ya:1 dan No:0)
- time -> Follow-up period (days)
- DEATH_EVENT -> If the patient died during the follow-up period (boolean) (Ya:1 dan No:0)

# Data Preprocessing

## Data Numerik

```python
#Cek Multicollinearity

from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data = pd.DataFrame()
vif_data["feature"] = num_dat.columns
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(num_dat.values, i) for i in range(len(num_dat.columns))]

vif_data
```

- Hanya ada 3 feature yang tidak saling berkaitan dengan feature lainnya yaitu creatinine_phosphokinase, serum_creatinine, dan time
- Dapat diprediksi bahwa jika dilakukan reduksi dimensi, maka paling tidak dapat direduksi sampai 4 dimensi

### Reduksi Dimensi Menggunakan PCA

**Tanpa Scaling Data**

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=7).fit(num_dat)
```

```python
pca_coba=PCA(n_components=0.8)
pca_coba.fit(num_dat)
```

```python
pca_coba.n_components_
```

```python
fig,ax=plt.subplots(ncols=2,figsize=(16,5))
ax[0].plot(range(1,8),pca.explained_variance_ratio_)
ax[0].set_xlabel('Komponen')
ax[0].set_ylabel('Explained Variance Ratio')

ax[1].plot(range(1,8),np.cumsum(pca.explained_variance_ratio_))
ax[1].set_xlabel('Jumlah Komponen')
ax[1].set_ylabel('Kumulatif Explained Var Ratio')
plt.show()
```

- Hanya komponen 1 yang memiliki explained variance ratio yang mendekati 1 sisanya mendekati nol
- Namun, jika dilihat dari cumulative explained variance rationya, jika hanya menggunakan 1 komponen, sangat tidak optimal karena cumsum EVR nya sangat kecil bahkan minus
- feature harus discale sebelum direduksi dimensinya dengan PCA

**Dengan Scaling**

```python
from sklearn.preprocessing import StandardScaler

num_scaler=StandardScaler()
num_dat_scaled=num_scaler.fit_transform(num_dat)
```

```python
from sklearn.decomposition import PCA
pca_scaled=PCA(n_components=7).fit(num_dat_scaled)
pca_scaled.explained_variance_ratio_
```

```python
fig,ax=plt.subplots(ncols=2,figsize=(16,5))
ax[0].plot(range(1,8),pca_scaled.explained_variance_ratio_)
ax[0].set_xlabel('Komponen')
ax[0].set_ylabel('Explained Variance Ratio')

ax[1].plot(range(1,8),np.cumsum(pca_scaled.explained_variance_ratio_))
ax[1].set_xlabel('Jumlah Komponen')
ax[1].set_ylabel('Kumulatif Explained Var Ratio')
plt.show()
```

- Nampak bahwa adanya degradasi nilai EVR yang gradual dari komponen pertama hingga terakhir
- Nampak ada patahan di 5 komponen, dengan cumsum EVR 0.8, kemungkinan 5 komponen merupakan jumlah yang terbaik dengan cumsum EVR sekitar 80%, namun 6 komponen terlihat bagus karena cumsum EVR nya mendekati 90% 

```python
pca_2d=PCA(n_components=2).fit(num_dat)
pca_2d_scaled=PCA(n_components=2).fit(num_dat_scaled)

twod=pca_2d.transform(num_dat)
twod_scaled=pca_2d_scaled.transform(num_dat_scaled)

fig,ax=plt.subplots(ncols=2,figsize=(16,5))
sns.scatterplot(x=twod[:,0],y=twod[:,1],hue=data.death,ax=ax[0])
ax[0].set_xlabel('PC 1')
ax[0].set_ylabel('PC 2')
ax[0].set_title('Unscaled Data')

sns.scatterplot(x=twod_scaled[:,0],y=twod_scaled[:,1],hue=data.death,ax=ax[1])
ax[1].set_xlabel('PC 1')
ax[1].set_ylabel('PC 2')
ax[1].set_title('Scaled Data')
plt.show()
```

Scaled data menunjukkan hasil yang terbaik, yaitu masih tampak pemisahan antar dua kelas. Meski demikian, perlu meninjau metode DR lainnya.

Berdasarkan plot kumulatif explained variance ratio (EVR), jumlah komponen/dimensi yang ideal adalah 5 - 6 dengan cumsum EVR 80 - 90%

### Reduksi Dimensi Menggunakan Kernel PCA

**Kernel Poly**

```python
from sklearn.decomposition import KernelPCA
```

```python
len(num_dat_scaled)
```

```python
kpca_7d=KernelPCA(n_components=7,kernel='poly').fit_transform(num_dat_scaled)

ev_kpca_7d = np.var(kpca_7d, axis=0)
evr_kpca_7d = ev_kpca_7d / np.sum(ev_kpca_7d)

fig,ax=plt.subplots(ncols=2,figsize=(16,5))
ax[0].plot(range(1,8),evr_kpca_7d)
ax[0].set_xlabel('Komponen')
ax[0].set_ylabel('Explained Variance Ratio')

ax[1].plot(range(1,8),np.cumsum(evr_kpca_7d))
ax[1].set_xlabel('Jumlah Komponen')
ax[1].set_ylabel('Kumulatif Explained Var Ratio')
plt.show()
```

cumsum EVR mengalami improvisasi dimana pada 4 komponen cumsum EVR di atas 80% dan 5 komponen di sekitar 90%

```python
kpca_2d=KernelPCA(n_components=2,kernel='poly').fit(num_dat_scaled)

twod=kpca_2d.transform(num_dat_scaled)

fig,ax=plt.subplots()
sns.scatterplot(x=twod[:,0],y=twod[:,1],hue=data.death)
ax.set_xlabel('PC 1')
ax.set_ylabel('PC 2')
ax.set_title('Kernel PCA - Poly')
plt.show()
```

Jika diplot 2 dimensi, kernel poly tidak terlihat pemisahan antar kelas yang jelas

**Kernel RBF**

```python
kpca_7d=KernelPCA(n_components=7,kernel='rbf').fit_transform(num_dat_scaled)

ev_kpca_7d = np.var(kpca_7d, axis=0)
evr_kpca_7d = ev_kpca_7d / np.sum(ev_kpca_7d)

fig,ax=plt.subplots(ncols=2,figsize=(16,5))
ax[0].plot(range(1,8),evr_kpca_7d)
ax[0].set_xlabel('Komponen')
ax[0].set_ylabel('Explained Variance Ratio')

ax[1].plot(range(1,8),np.cumsum(evr_kpca_7d))
ax[1].set_xlabel('Jumlah Komponen')
ax[1].set_ylabel('Kumulatif Explained Var Ratio')
plt.show()
```

Kernel rbf menunjukkan hasil yang tidak jauh berbeda dengan PCA linear dari segi cumulative EVR

```python
kpca_2d=KernelPCA(n_components=2,kernel='rbf').fit(num_dat_scaled)

twod=kpca_2d.transform(num_dat_scaled)

fig,ax=plt.subplots()
sns.scatterplot(x=twod[:,0],y=twod[:,1],hue=data.death)
ax.set_xlabel('PC 1')
ax.set_ylabel('PC 2')
ax.set_title('Kernel PCA - RBF')
plt.show()
```

Kernel RBF menunjukkan adanya pemisahan antar kelas, namun masih banyak yang saling tumpang tindih

**Kernel Sigmoid**

```python
kpca_7d=KernelPCA(n_components=7,kernel='sigmoid').fit_transform(num_dat_scaled)

ev_kpca_7d = np.var(kpca_7d, axis=0)
evr_kpca_7d = ev_kpca_7d / np.sum(ev_kpca_7d)

fig,ax=plt.subplots(ncols=2,figsize=(16,5))
ax[0].plot(range(1,8),evr_kpca_7d)
ax[0].set_xlabel('Komponen')
ax[0].set_ylabel('Explained Variance Ratio')

ax[1].plot(range(1,8),np.cumsum(evr_kpca_7d))
ax[1].set_xlabel('Jumlah Komponen')
ax[1].set_ylabel('Kumulatif Explained Var Ratio')
plt.show()
```

Improvisasi cumulative EVR dari PCA hanya terjadi pada Kernel PCA dengan kernel poly namun dari segi pemisahan antar kelas, lebih bagus kernel rbf meski pada jumlah komponen yang sama, cumsum EVR nya lebih kecil daripada yang lain

### Reduksi Dimensi Menggunakan Manifold Learning - T-SNE

TSNE menggunakan metode yang mirip dengan KMeans dimana model akan menghitung jarak antar data dan antar distribusi sehingga metrik yang digunakan untuk mengevaluasi bukan explained_variance_ratio melainkan Kullback-Leibler Divergence (Meminimalkan KL Divergence)

```python
from sklearn.manifold import TSNE

tsne_kl=[]
for i in range(1,8):
    tsne_kl.append(TSNE(n_components=i,method='exact').fit(num_dat_scaled).kl_divergence_)


fig,ax=plt.subplots(figsize=(5,5))
ax.plot(range(1,8),tsne_kl)
ax.set_xlabel('Komponen')
ax.set_ylabel('Kullback-Leibler Divergence')
plt.show()
```

Jumlah komponen 2 menunjukkan nilai Kullback-Leibler paling kecil di antara yang lain

# Mana yang paling baik di antara banyak metode?

Kita perlu mengecek dari salah satu metrik yaitu akurasi dengan model yang sama dan dengan jumlah komponen yang sama yaitu 5

```python
#Transformasi data numerik dengan semua metode yang sudah ditinjau
X_no_DR=data.drop(columns='death')
X_no_DR_mcol=data.drop(columns=['age','ejection_fraction','platelets','serum_sodium','death'])
X_pca=PCA(n_components=5).fit_transform(num_dat_scaled)
X_kpca_poly=KernelPCA(n_components=5,kernel='poly').fit_transform(num_dat_scaled)
X_kpca_rbf=KernelPCA(n_components=5,kernel='rbf').fit_transform(num_dat_scaled)
X_kpca_sigmoid=KernelPCA(n_components=5,kernel='sigmoid').fit_transform(num_dat_scaled)
X_tsne_10=TSNE(n_components=2,perplexity=10,method='exact').fit_transform(num_dat_scaled)
X_tsne_30=TSNE(n_components=2,perplexity=30,method='exact').fit_transform(num_dat_scaled)
X_tsne_50=TSNE(n_components=2,perplexity=50,method='exact').fit_transform(num_dat_scaled)
X_tsne_100=TSNE(n_components=2,perplexity=100,method='exact').fit_transform(num_dat_scaled)
X_isomap=Isomap(n_components=5).fit_transform(num_dat_scaled)
```

```python
X_pca=np.concatenate((X_pca,cat_dat),axis=1)
X_kpca_poly=np.concatenate((X_kpca_poly,cat_dat),axis=1)
X_kpca_rbf=np.concatenate((X_kpca_rbf,cat_dat),axis=1)
X_kpca_sigmoid=np.concatenate((X_kpca_sigmoid,cat_dat),axis=1)
X_tsne_10=np.concatenate((X_tsne_10,cat_dat),axis=1)
X_tsne_30=np.concatenate((X_tsne_30,cat_dat),axis=1)
X_tsne_50=np.concatenate((X_tsne_50,cat_dat),axis=1)
X_tsne_100=np.concatenate((X_tsne_100,cat_dat),axis=1)
X_isomap=np.concatenate((X_isomap,cat_dat),axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

**Random Forest**

```python
methods=['No-DR','Multicolinearity','PCA','KPCA Poly','KPCA RBF','KPCA Sigmoid','TSNE p-10','TSNE p-30','TSNE p-50','TSNE p-100','Isomap']
X=[X_no_DR,X_no_DR_mcol,X_pca, X_kpca_poly, X_kpca_rbf, X_kpca_sigmoid, X_tsne_10, X_tsne_30, X_tsne_50, X_tsne_100, X_isomap]

acc=[]
auc=[]
acc_methods=pd.DataFrame(index=methods)

for i in range(len(methods)):
    X_train,X_test,y_train,y_test=train_test_split(X[i],data['death'],test_size=0.3,stratify=data['death'],random_state=50)
    scaler=StandardScaler()
    X_train=scaler.fit_transform(X_train)
    
    model=RandomForestClassifier()
    model.fit(X_train,y_train)
    
    X_test=scaler.transform(X_test)
    y_pred=model.predict(X_test)
    
    acc.append(accuracy_score(y_test,y_pred))
    auc.append(roc_auc_score(y_test,model.predict_proba(X_test)[:,1]))

acc_methods['Accuracy']=acc
acc_methods['AUC Score']=auc

acc_methods.plot(kind='bar')
acc_methods
```

**SVM**

```python
methods=['No-DR','Multicolinearity','PCA','KPCA Poly','KPCA RBF','KPCA Sigmoid','TSNE p-10','TSNE p-30','TSNE p-50','TSNE p-100','Isomap']
X=[X_no_DR,X_no_DR_mcol,X_pca, X_kpca_poly, X_kpca_rbf, X_kpca_sigmoid, X_tsne_10, X_tsne_30, X_tsne_50, X_tsne_100, X_isomap]

acc=[]
auc=[]
acc_methods=pd.DataFrame(index=methods)

for i in range(len(methods)):
    X_train,X_test,y_train,y_test=train_test_split(X[i],data['death'],test_size=0.3,stratify=data['death'],random_state=50)
    scaler=StandardScaler()
    X_train=scaler.fit_transform(X_train)
    
    model=SVC(probability=True)
    model.fit(X_train,y_train)
    
    X_test=scaler.transform(X_test)
    y_pred=model.predict(X_test)
    
    acc.append(accuracy_score(y_test,y_pred))
    auc.append(roc_auc_score(y_test,model.predict_proba(X_test)[:,1]))

acc_methods['Accuracy']=acc
acc_methods['AUC Score']=auc

acc_methods.plot(kind='bar')
acc_methods
```
